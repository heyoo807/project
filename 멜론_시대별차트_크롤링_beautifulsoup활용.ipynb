```python
from bs4 import BeautifulSoup as bs
import requests as req
header = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'}
```

```python
# ajax로 받은 url 안 문자 변경을 반복해서 시대별 차트 제목 크롤링

song_titles = {}
for year in range(1985, 2023):  # 1985부터 2023까지 년도 변경
    base_url = 'https://www.melon.com/chart/age/list.htm?idx={}&chartType=YE&chartGenre=KPOP&chartDate={}&moved=Y'
    for idx in [1, 51]:  # idx는 1 또는 51
        print(idx)
        print(year)
        if idx == 1:
          song_titles[year] = []
        url = base_url.format(idx, year)
        res = req.get(url, headers=header)
        html = bs(res.text)
        intag = html.select('.ellipsis.rank01>span>strong>a')
        song_titles[year] =  [i.text for i in intag]
```

```python
import pandas as pd
# 딕셔너리를 데이터프레임으로 변환
df = pd.DataFrame.from_dict(song_titles,  orient='index')

# 데이터프레임을 엑셀 파일로 저장
df = df.transpose()
df.head()
df.to_excel('song_titles.xlsx', index = False, encoding = 'utf-8')
```
